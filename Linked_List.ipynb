# Linked List : Creation

```python
class Node:
  def __init__(self,data):
    self.data = data
    self.next = None

a = Node(5)
b = Node(8)
c = Node(10)

a.next = b
b.next = c
head = a

print(head.data)
print(head.next.data)
print(head.next.next.data)
```

## List Traversal

```python
def list_traversal(head):
  curr = head
  while curr!=None:
    print(curr.data)
    curr = curr.next

list_traversal(head)
```

## Insertion

### Insertion at the beginning of the linked list

```python
new = Node(3)
def insertion_at_head(newNode, head):
  newNode.next = head
  head = newNode
  return head


head = insertion_at_head(new, head)
list_traversal(head)
```

###Insertion At the End of the linked List


```python
newNode = Node(15)
def insert_at_tail(head, newNode):
  curr = head
  while curr.next != None:
    curr = curr.next

  curr.next = newNode
  return curr


insert_at_tail(head, newNode)
list_traversal(head)
```

###Insertion at the Kth Index

```python
def insert_at_kth(head, newNode, k):
  curr = head
  for i in range(k-1):
    curr = curr.next # 3 5 8 10 15

  newNode.next = curr.next
  curr.next = newNode

newNode = Node(6)
k = 2

insert_at_kth(head, newNode, k)
list_traversal(head)
```

## Deletion

###Deletion of the first node

```python
head = head.next
list_traversal(head)
```

### Deletion of the last node

```python
curr = head
while curr.next.next != None:
  curr = curr.next

curr.next = None

list_traversal(head)
```

### Delete at kth

```python
def delete_at_kth(head, k):
  for i in range(k-1):
    head = head.next

  head.next = head.next.next

delete_at_kth(head, 2)
list_traversal(head)
```

# Doubly Linked List

```python
class doublyLL:
  def __init__(self,data):
    self.data = data
    self.next = None
    self.prev = None

a = doublyLL(10)
b = doublyLL(20)
c = doublyLL(30)
d = doublyLL(40)

a.next = b
b.next = c
c.next = d
d.prev = c
c.prev = b
b.prev = a
```

# Circular Linked List

```python
class Node:
  def __init__(self,data):
    self.data = data
    self.next = None

a = Node(5)
b = Node(8)
c = Node(10)

a.next = b
b.next = c
c.next = a
head = a

print(head.data)
print(head.next.data)
print(head.next.next.data)
print(head.next.next.next.data)
```

```python
curr = head
while True:
  print(curr.data, end = " ")
  curr = curr.next
  if curr == head:
    break
```
