# __init__ and self keywords

```python
class Food:
  def __init__(self,type, cals):
    self.type = type
    self.cals = cals

  def food_info(self):
    print(f"It is a {self.type} and it has {self.cals} calories")

  def compare(self, other):
    if self.cals < other.cals:
      print(f"{self.cals} is a good food for weight loss")
    elif self.cals > other.cals:
      print(f"{self.cals} is a good food for weight gain")
    else:
      print("Both are same")

Burger = Food("Junk Food", 300)
Salad = Food("Healthy Food", 100)
Food.food_info(Burger)

Burger.cals = 400
Burger.food_info()
Burger.compare(Salad)
```

```python
class student:
  # static variable
  school = "All these student records belongs to Babu Banarsi Das University"
  # constructor method
  def __init__(self):
    self.name = None
    self.year = None
    self.course = None
    self.m1 = None
    self.m2 = None
    self.m3 = None

  #Instance Method

  #Mutator
  def set_info(self,name,year,course,m1,m2,m3):
    self.name = name
    self.year = year
    self.course = course
    self.m1 = m1
    self.m2 = m2
    self.m3 = m3

  #Accessor
  def get_info(self):
    print("Name :", self.name)
    print("Year :", self.year)
    print("Course :", self.course)
    print("Maths marks :", self.m1)
    print("DSA marks :", self.m2)
    print("Machine Learning marks :", self.m3)

  def get_avg(self):
    print("Average marks :",(self.m1+self.m2+self.m3)/3)


  #class method
  @classmethod
  def school_info(cls):
    print(cls.school)


  #static method
  @staticmethod
  def student_info():
    print("All students have scored good marks")



Dishambha = student()
Dishambha.set_info("Dishambha", 4, "B.tech CSE", 99,98,97)
Dishambha.get_info()
Dishambha.get_avg()
student.school_info()
student.student_info()


```

# INHERITANCE


## Single LeveL Inheritance

```python
class A():
  def feature1(self):
    print("Feature 1 is working")
  def feature2(self):
    print("Feature 2 is working")

class B(A):
  def feature3(self):
    print("Feature 3 is working")
  def feature4(self):
    print("Feature 4 is working")

obj1 = A()
obj1.feature1()
obj1.feature2()
print(" ")
obj2 = B()
obj2.feature1()
obj2.feature2()
obj2.feature3()
obj2.feature4()
```

## Multilevel Inheritance

```python
class A():
  def feature1(self):
    print("Feature 1 is working")
  def feature2(self):
    print("Feature 2 is working")

class B(A):
  def feature3(self):
    print("Feature 3 is working")
  def feature4(self):
    print("Feature 4 is working")

class C(B):
  def feature5(self):
    print("Feature 5 is working")

print("Object 1")
obj1 = A()
obj1.feature1()
obj1.feature2()
print(" ")
print("Object 2")
obj2 = B()
obj2.feature1()
obj2.feature2()
obj2.feature3()
obj2.feature4()
print(" ")
print("Object 3")
obj3 = C()
obj3.feature1()
obj3.feature2()
obj3.feature3()
obj3.feature4()
obj3.feature5()
```

## Multiple Inheritance

```python
class A:
  def feature1(self):
    print("Feature 1 is working")
  def feature2(self):
    print("Feature 2 is working")

class B:
  def feature3(self):
    print("Feature 3 is working")
  def feature4(self):
    print("Feature 4 is working")

class C(A,B):
  def feature5(self):
    print("Feature 5 is working")

print("Object 1")
obj1 = A()
obj1.feature1()
obj1.feature2()
print(" ")
print("Object 2")
obj2 = B()
obj2.feature3()
obj2.feature4()
print(" ")
print("Object 3")
obj3 = C()
obj3.feature1()
obj3.feature2()
obj3.feature3()
obj3.feature4()
obj3.feature5()
```

## Hierarchical Inheritance

```python
class Parent:
  def house1(self):
    print("Parents have 1 house")

class son(Parent):
  def house2(self):
    print("Son has a new house")

class daughter(Parent):
  def house3(self):
    print("Daughter has a new house")

baap = Parent()
launda = son()
laundia = daughter()

baap.house1()
print("")
launda.house1()
launda.house2()
print()
laundia.house1()
laundia.house3()
```

# Duck Typing

* Duck typing is a form of polymorphism, especially in dynamically typed languages like Python, where an object's type is less important than the methods it possesses

```python
class PyCharm:
  def execute(self):
    print("compiling...")
    print("executing...")
    print("//CODE//")
    print("Completed")

class VsCode:
  def execute(self):
    print("type checking..")
    print("compiling...")
    print("executing...")
    print("//CODE//")
    print("Completed")

class Laptop:
  def code(self,ide):
    ide.execute()

ide = PyCharm()

# Create an instance of Laptop
my_laptop = Laptop()
# Call the code method on the instance
my_laptop.code(ide)

ide = VsCode()
my_lap2 = Laptop()
print("")
my_lap2.code(ide)
```

# Operator Overloading

```python
class Studentt:
  def __init__(self, m1, m2):
    self.m1 = m1
    self.m2 = m2

  def __add__(self, other):
    m1 = self.m1 + other.m1
    m2 = self.m2 + other.m2
    s3 = Studentt(m1, m2)
    return s3

  def __gt__(self, other):
    s1 = self.m1 + self.m2
    s2 = other.m1 + other.m2
    if s1 > s2:
      return True
    else :
      return False

  def __str__(self):
    return '{} {} '.format(self.m1, self.m2)





s1 = Studentt(58, 83)
s2 = Studentt(99, 73)

s3 = s1 + s2

print(f"Student 3 marks: m1={s3.m1}, m2={s3.m2}")


if s1 > s2:
  print("s1 wins")
else:
  print("s2 wins")


print(s1)
print(s2)
```

# Method Overloading and Method Overriding


## Overloading

```python
class Studentt:
  def __init__(self, m1, m2):
    self.m1 = m1
    self.m2 = m2

  def sum(self, a=None,b=None,c=None):
    s = 0
    if a != None and b != None and c != None:
      s = a + b + c
    elif a!=None and b!=None:
      s = a + b
    else:
      s = a
    return s

s1 = Studentt(58, 83)
print(s1.sum(2,4,9))
```

# Object-Oriented Programming System (OOPS)

## What is OOPS?
* Object-Oriented Programming System (OOPS) is a programming paradigm based on the concept of objects.
* Objects contain data (variables) and methods (functions).
* OOPS improves code reusability, modularity, scalability, and security.

## Core Concepts of OOPS

### Class
* A class is a blueprint or template used to create objects.
* It defines attributes (data members) and methods (member functions).
* Example: A Car class may contain attributes like color, speed and methods like drive().

### Object
* An object is an instance of a class.
* It represents a real-world entity.
* Example: myCar is an object of the Car class.

### Encapsulation
* Encapsulation is the process of wrapping data and methods together into a single unit.
* It prevents direct access to data and protects it from unauthorized modification.
* Achieved using access specifiers:
  * public
  * private
  * protected
* Advantage: Improves data security and control.

### Abstraction
* Abstraction means hiding implementation details and showing only essential features.
* It focuses on what an object does rather than how it does it.
* Implemented using:
  * abstract classes
  * interfaces
* Example: Using a mobile phone without knowing its internal hardware.

### Inheritance
* Inheritance allows one class to acquire properties and methods of another class.
* It promotes code reusability.
* Types of inheritance:
  * Single
  * Multilevel
  * Hierarchical
* Multiple inheritance depends on language support.
* Example: ElectricCar inherits from Car.

### Polymorphism
* Polymorphism means one function, many forms.
* Same method name with different behavior.
* Types of polymorphism:
  * Compile-time polymorphism (Method Overloading)
  * Run-time polymorphism (Method Overriding)
* Example: add(int, int) and add(int, int, int)

## Advantages of OOPS
* Code reusability
* Data security
* Easy maintenance
* Scalability
* Real-world modeling

## Disadvantages of OOPS
* Complex design
* Higher memory usage
* Slower execution compared to procedural programming

* Encapsulation: Binding data and methods together
* Abstraction: Hiding internal details
* Inheritance: Acquiring properties of another class
